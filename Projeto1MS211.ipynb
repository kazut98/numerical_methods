```python
# Pacotes 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Projeto I Questao a
def f_a(x): 
  return x**3 - x**(-1) + 1

def df_a(x):
  return 3*x**2 + x**(-2)

def metodonewton_a(xo, niteracoes, Epsilon, lista):
  x = xo
  i = 0
  E = 1
  while(E >= Epsilon and i < niteracoes):
    xn = x
    x = xn - f_a(xn)/df_a(xn)
    E = np.abs((x-xn/x))
    lista.append(x)
    i = i + 1
  
  return (x, i, E, lista)

```

```python
#iterações com chute inicial = 1.5 na questão a 
xo = 1.5
Epsilon = 10**(-6)
niteracoes = 5
lista = []
resposta = metodonewton_a(xo, niteracoes, Epsilon, lista)

print('raiz = ', resposta[0])
print('numero de iteracoes = ', resposta[1])
print(resposta[3])

#tabela e gráfico dos valores das iterações na questão a 
xa = resposta[3]
ya = [xa[0]**3 - xa[0]**(-1) + 1, xa[1]**3 - xa[1]**(-1) + 1,xa[2]**3 - xa[2]**(-1) + 1,xa[3]**3 - xa[3]**(-1) + 1, xa[4]**3 - xa[4]**(-1) + 1]
plt.scatter(xa, ya)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Projeto I Questão b

pi = 3.141592
def func_b(x):
  return pi - (1/x)
def deriv_b(x):
  return x**(-2)

def newton_b(xo_b, niteracoes_b, Epsilon_b, lista_b):
  x = xo_b
  i = 0
  E = 1
 
  while(E >= Epsilon_b and i < niteracoes_b):
    xn = x
    x = xn - func_b(xn)/deriv_b(xn)
    E = np.abs((x-xn/x))
    lista_b.append(x)
    i = i + 1
  
  return (x, i, E, lista_b)
  
```

```python
#com chute inicial = 0.5 na questão b
xo_b1 = 0.5
Epsilon_b1 = 10**(-6)
niteracoes_b1 = 5
lista_b1 = []
resposta_b1= newton_b(xo_b1, niteracoes_b1, Epsilon_b1, lista_b1)

print('raiz = ', resposta_b1[0])
print('numero de iteracoes = ', resposta_b1[1])
print(resposta_b1[3])

#tabela e gráfico dos valores das iterações
xa_b1 = resposta_b1[3]
ya_b1 = [pi - (1/xa_b1[0]), pi - (1/xa_b1[1]), pi - (1/xa_b1[2]), pi - (1/xa_b1[3]), pi - (1/xa_b1[4])]
plt.plot(xa_b1, ya_b1)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
# Projeto I Questao c1
def f_c1(x):
  return x**(1/3)

def df_c1(x):
  return (1/3)*x**(-2/3)

def metodonewton_c1(xo, niteracoes, Epsilon, lista):
  x = xo
  i = 0
  E = 1
 
  while(E >= Epsilon and i < niteracoes):
    xn = x
    x = xn - f_c1(xn)/df_c1(xn)
    E = np.abs((x-xn/x))
    lista.append(x)
    i = i + 1
  
  return (x, i, E, lista)
```

```python
#iterações com chute inicial = 0.5 na questão c1
xo = 0.5
Epsilon = 10**(-5)
niteracoes = 5
lista = []
resposta = metodonewton_c1(xo, niteracoes, Epsilon, lista)

print('raiz = ', resposta[0])
print('numero de iteracoes = ', resposta[1])
print(resposta[3])

#gráfico dos valores das iterações da questão c1
xa = resposta[3]
ya = [xa[0]**(1/3),xa[1]**(1/3),xa[2]**(1/3),xa[3]**(1/3),xa[4]**(1/3)]
plt.plot(xa, ya)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#questão c2 

def func_c(x):
  return (x**3) - (5*x)
def deriv_c(x):
  return (3*x**2) - 5

def newton_c(xo_c, niteracoes_c, Epsilon_c, lista_c):
  x = xo_c
  i = 0
  E = 1
 
  while(E >= Epsilon_c and i < niteracoes_c):
    xn = x
    x = xn - func_c(xn)/deriv_c(xn)
    E = np.abs((x-xn/x))
    lista_c.append(x)
    i = i + 1
  
  return (x, i, E, lista_c)
```
