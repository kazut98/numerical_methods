```python
# Pacotes 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Projeto I Questao a
def f(x): 
  return x**3 - x**(-1) + 1

def df(x):
  return 3*x**2 + x**(-2)

def metodonewton(xo, niteracoes, Epsilon, lista):
  x = xo
  i = 0
  E = 1
  while(E >= Epsilon and i < niteracoes):
    xn = x
    x = xn - f(xn)/df(xn)
    E = np.abs((x-xn/x))
    lista.append(x)
    i = i + 1
  
  return (x, i, E, lista)

xo = 1.5
Epsilon = 10**(-6)
niteracoes = 5
lista = []
resposta = metodonewton(xo, niteracoes, Epsilon, lista)

print('raiz = ', resposta[0])
print('numero de iteracoes = ', resposta[1])
print(resposta[3])

#tabela e gráfico dos valores das iterações
xa = resposta[3]
ya = [xa[0]**3 - xa[0]**(-1) + 1, xa[1]**3 - xa[1]**(-1) + 1,xa[2]**3 - xa[2]**(-1) + 1,xa[3]**3 - xa[3]**(-1) + 1, xa[4]**3 - xa[4]**(-1) + 1]
plt.scatter(xa, ya)
plt.show()
```

```python
# Projeto I Questão b
 #ANOTAÇÃO: tomar cuidado com variáveis 
pi = 3.141592
def func(x):
  return pi - (1/x)
def deriv(x):
  return x**(-2)

def newton_b(xo_b, niteracoes_b, Epsilon_b, lista_b):
  x = xo_b
  i = 0
  E = 1
 
  while(E >= Epsilon_b and i < niteracoes_b):
    xn = x
    x = xn - func(xn)/deriv(xn)
    E = np.abs((x-xn/x))
    lista_b.append(x)
    i = i + 1
  
  return (x, i, E, lista_b)

#com chute inicial = 0.5 
xo_b1 = 0.5
Epsilon_b1 = 10**(-6)
niteracoes_b1 = 5
lista_b1 = []
resposta_b1= newton_b(xo_b1, niteracoes_b1, Epsilon_b1, lista_b1)

print('raiz = ', resposta_b1[0])
print('numero de iteracoes = ', resposta_b1[1])
print(resposta_b1[3])

#com chute inicial = 0.7
xo_b2 = 0.7
Epsilon_b2 = 10**(-6)
niteracoes_b2 = 5
lista_b2 = []
resposta_b2 = newton_b(xo_b2, niteracoes_b2, Epsilon_b2, lista_b2)

print('raiz = ', resposta_b2[0])
print('numero de iteracoes = ', resposta_b2[1])
print(resposta_b2[3])
```

```python
#tabela e gráfico dos valores das iterações para 0.5
xa_b1 = resposta_b1[3]
ya_b1 = [pi - (1/xa_b1[0]), pi - (1/xa_b1[1]), pi - (1/xa_b1[2]), pi - (1/xa_b1[3]), pi - (1/xa_b1[4])]
plt.plot(xa_b1, ya_b1)
plt.show()
```
